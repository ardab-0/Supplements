### 7.3 Programming Task: Color Quantization using K-Means clustering.

 The goal of this task is to perform color quantization using the K-Means clustering algorithm.

   i\. Implement the K-Means clustering algorithm using Numpy/SciPy.




```python
def K_means_clustering(train_X, k):
   centers = np.random.uniform( size=(k, len(train_X[0])) )
   centers *= 256
   centers = np.floor(centers)
   #TODO: Implement the K means algorithm. The input train_X is an array of shape (p,3), where p is the number of pixels in the image.
   i = 0
   while True:
      print(i)
      clusters = pairwise_distances_argmin(train_X, centers)
      centroids_update = np.array([train_X[clusters == i].mean(axis=0) for i in range(k)])
      centroids_update = np.nan_to_num(centroids_update)
      if np.all(centers == centroids_update):
         break 
      centers = centroids_update
      i+=1
   return centers
   # return cluster_centers

```

   ii\. Read the given image. Consider each pixel as a 3-dimensional vector and run your K-Means clustering algorithm to get 64 clusters.


```python
#TODO
im = skimage.io.imread("sup_7_3_input.jpg")
im_shape = im.shape
k = 64
pixels = im.copy()
pixels = pixels.reshape((-1, 3))

cluster_centers = K_means_clustering(pixels, k=k)
plt.scatter(pixels[:, 0], pixels[:, 1], s=40, cmap='gray')
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], c='red', s=40, alpha=0.8)
plt.plot()
```
